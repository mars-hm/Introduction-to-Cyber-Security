```python
#DES Encryption

import pyDes

def encryption(key, plaintext):
    des = pyDes.des(key, pad=None, padmode=pyDes.PAD_PKCS5)
    cipher = des.encrypt(plaintext)
    return cipher

def decryption(key, ciphertext):
    des = pyDes.des(key, pad=None, padmode=pyDes.PAD_PKCS5)
    plain = des.decrypt(ciphertext)
    return plain

plaintext = input("Enter the Plaintext: ").encode()
#Baboons are cuties.
key = b'blehblee'
ciphertext = encryption(key, plaintext)
print("Ciphertext: ", ciphertext.hex())

plaint = decryption(key, ciphertext)
print("Decrypted Plaintext: ", plaint.decode())
```

```python
#AES

from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend
from cryptography.hazmat.primitives import padding

def encryption(key, iv, plaintext):
    cipher = Cipher(algorithms.AES(key), modes.CBC(iv))
    encryptor = cipher.encryptor()
    padder = padding.PKCS7(128).padder()
    padded_plaintext = padder.update(plaintext) + padder.finalize()
    ciphertext = encryptor.update(padded_plaintext) + encryptor.finalize()
    return ciphertext

def decryption(key, iv, ciphertext):
    cipher = Cipher(algorithms.AES(key), modes.CBC(iv))
    decryptor = cipher.decryptor()
    padded_plaintext = decryptor.update(ciphertext) + decryptor.finalize()
    unpadder = padding.PKCS7(128).unpadder()
    plaintext = unpadder.update(padded_plaintext) + unpadder.finalize()
    return plaintext

plaintext = input("Enter the Plaintext: ").encode()
#Baboons are different from Bumble Bees.
key = b'blahbleeblehblue'  # 16-byte key for AES-128
iv = b'bippitybopityboo'  # 16-byte IV (Initialization Vector)
ciphertext = encryption(key, iv, plaintext)
print("Ciphertext:", ciphertext.hex())

plaint = decryption(key, iv, ciphertext)
print("Decrypted Plaintext: ", plaint.decode())
```

```python
#RSA

import random
import sympy
from sympy import mod_inverse

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def is_prime(num, k=10):
    return sympy.isprime(num)

def generate_prime(bits):
    num = random.getrandbits(bits)
    while not is_prime(num):
        num = random.getrandbits(bits)
    return num

def generate_keypair(bits):
    # Generate one distinct prime number p
    p = generate_prime(bits)

    # Find another prime number q close to p
    q = generate_prime(bits)
    while abs(p - q) < 2**((bits-1)//2):
        q = generate_prime(bits)

    # Calculate n (modulus)
    n = p * q

    # Calculate Euler's totient function (φ)
    phi = (p - 1) * (q - 1)

    # Choose an integer e such that 1 < e < φ(n) and gcd(e, φ(n)) = 1
    e = random.randrange(2**16, phi)
    while gcd(e, phi) != 1:
        e = random.randrange(2**16, phi)

    # Calculate the modular multiplicative inverse of e modulo φ(n)
    d = mod_inverse(e, phi)

    # Public key: (e, n), Private key: (d, n)
    public_key = (e, n)
    private_key = (d, n)

    return public_key, private_key

def encrypt(message, public_key):
    e, n = public_key
    cipher = []
    for char in message:
      encrypted_char = pow(ord(char), e, n)
      cipher.append(encrypted_char)
    return cipher

def decrypt(encrypted_message, private_key):
    d, n = private_key
    plain = []
    for char in encrypted_message:
      decrypted_char = chr(pow(char, d, n))
      plain.append(decrypted_char)
    return ''.join(plain)

# Baboons are different from Bamboos.
plaintext = input("Enter the Plaintext: ")

public_key, private_key = generate_keypair(1024)

cipher = encrypt(plaintext, public_key)

dplain = decrypt(cipher, private_key)

print("Original message: ", plaintext)
print("Ciphertext: ", cipher)
print("Decrypted Plaintext: ", dplain)
```

```python
def mod_exp(base, exponent, modulus):
    result = 1
    while exponent > 0:
        if exponent % 2 == 1:
            result = (result * base) % modulus
        exponent //= 2
        base = pow(base, 2, modulus)
    return result

def generate_shared_secret(private_key, public_key, prime):
    shared_secret = mod_exp(public_key, private_key, prime)
    return shared_secret

def encrypt(message, shared_secret):
  cipher = ''
  for char in message:
    encrypted_char = chr(ord(char) ^ shared_secret)
    cipher += encrypted_char
  return cipher

def decrypt(encrypted_message, shared_secret):
    plain = ""
    for char in encrypted_message:
      decrypted_char = chr(ord(char) ^ shared_secret)
      plain += decrypted_char
    return plain

# Prime number and base
prime = 23
base = 5

# Alice's private key and public key
pvtkey_A = 6
pubkey_A = mod_exp(base, pvtkey_A, prime)

# Bob's private key and public key
pvtkey_B = 15
pubkey_B = mod_exp(base, pvtkey_B, prime)

# Shared secret computation
shared_secret_A = generate_shared_secret(pvtkey_A, pubkey_B, prime)
shared_secret_B = generate_shared_secret(pvtkey_B, pubkey_A, prime)

message = input("Enter the Plaintext: ")
#Bamboos are also different from Bumble Bees
encrypted_message = encrypt(message, shared_secret_A)
decrypted_message = decrypt(encrypted_message, shared_secret_B)

print("Original Message: ", message)
print("Ciphertext: ", encrypted_message)
print("Plaintext: ", decrypted_message)
```
